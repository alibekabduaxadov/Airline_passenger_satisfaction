```python
#yozuvli ustunlar xilmaxilligini ko'ramiz
str_col=['Gender','Customer Type','Type of Travel','Class']
for i in str_col:
  print(train_dataset[i].value_counts())
```

```python
numeric_col=['Age', 'Flight Distance', 'Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']
```

```python
#modellarni baholash uchun train va test qismlarga bo'lamiz
train_set, test_set = train_test_split(train_dataset, test_size=0.2, random_state=42)
```

```python
x=train_set.drop('satisfaction',axis=1)
y=train_set['satisfaction'].copy()
```

```python
numeric_pipline=Pipeline([
     ('scaler',StandardScaler())])
```

```python
#train datani full piplinedan o'tqazamiz
full_pipeline = ColumnTransformer([
    ('num', numeric_pipline, numeric_col),
    ('cat', OneHotEncoder(handle_unknown='ignore'), str_col)
])
train_pre = full_pipeline.fit_transform(x)
```

# ML ni boshlaymiz

**DecisionTreeClassifier modeli**

```python
# test datalarni piplinedan o'tqazamiz
x_test=test_set.drop(['satisfaction'],axis=1)
y_test=test_set['satisfaction']


x_test = full_pipeline.fit_transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
DT_model = DecisionTreeClassifier()

DT_model.fit(train_pre, y)

y_pred = DT_model.predict(x_test)

print(classification_report(y_test, y_pred) )
```

**RandomForestClassifier modeli**

```python
from sklearn.ensemble import RandomForestClassifier

RF_model = RandomForestClassifier()

RF_model.fit(train_pre, y)

y_pred = RF_model.predict(x_test)

print( classification_report(y_test, y_pred) )
```
